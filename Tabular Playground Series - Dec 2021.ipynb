```python
#making new columns

#Distance To Hydrology
train['Distance_To_Hydrology'] = (train['Horizontal_Distance_To_Hydrology']**2 + train['Vertical_Distance_To_Hydrology']**2)**0.5
test['Distance_To_Hydrology'] = (test['Horizontal_Distance_To_Hydrology']**2 + test['Vertical_Distance_To_Hydrology']**2)**0.5

#Absolute distance To Hydrology
train['Absolute_distance_To_Hydrology'] = np.abs(train['Horizontal_Distance_To_Hydrology']) + np.abs(train['Vertical_Distance_To_Hydrology'])
test['Absolute_distance_To_Hydrology'] = np.abs(test['Horizontal_Distance_To_Hydrology']) + np.abs(test['Vertical_Distance_To_Hydrology'])

#Wilderness_Area total
train['Wilderness_Area_total'] = train['Wilderness_Area1'] + train['Wilderness_Area2'] + train['Wilderness_Area3'] + train['Wilderness_Area4']
test['Wilderness_Area_total'] = test['Wilderness_Area1'] + test['Wilderness_Area2'] + test['Wilderness_Area3'] + test['Wilderness_Area4']
```

```python
#import packages for modelling
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#label encoding
le = LabelEncoder()
train["Cover_Type"] = le.fit_transform(train["Cover_Type"])
```

```python
#splitting train features and target
X = train.drop(['Cover_Type'],axis=1)
y = train['Cover_Type']
```

```python
#splitting to train and validation data
X_train, X_val, y_train, y_val = train_test_split(X, 
                                                  y, 
                                                  test_size=0.2, 
                                                  shuffle=False)
#number of feature
n_features = X_train.shape[1]
num_class = train["Cover_Type"].nunique()

print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
#define model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(512, activation='relu', input_shape=(n_features,)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(num_class, activation='softmax')
])

model.summary()
```

```python
#compile model
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#define callbacks
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('val_accuracy')/logs.get('accuracy')>0.95) and (logs.get('val_accuracy')>0.95):
      print("cancelling training!")
      self.model.stop_training = True

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=5)

early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=10,
    restore_best_weights=True)

callbacks = [myCallback(),reduce_lr, early_stop]
```

```python
#model fitting
history = model.fit(X_train,
                   y_train,
                   validation_data=(X_val,y_val),
                   epochs=50,
                   callbacks=callbacks,
                   batch_size=1024)
```

```python
#training plot
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(acc))
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy') 
plt.legend()
plt.figure()
```

```python
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

```python
#predict
test_predict = model.predict(test)
test_predict = np.argmax(test_predict, axis=1)
test_predict = le.inverse_transform(test_predict)
test_predict
```

```python
#creating submission
tabular_des = pd.DataFrame({'Id':sample_sub.Id, 'Cover_Type': test_predict})
tabular_des
```
