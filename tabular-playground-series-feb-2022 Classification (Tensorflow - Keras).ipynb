```python
#improting dataset
train = pd.read_csv('./Dataset/tabular-playground-series-feb-2022/train.csv')
test = pd.read_csv('./Dataset/tabular-playground-series-feb-2022/test.csv')
sample_sub = pd.read_csv('./Dataset/tabular-playground-series-feb-2022/sample_submission.csv')
```

```python
#dropping unsecessary column for model
train = train.drop(['row_id'],axis=1)
test = test.drop(['row_id'],axis=1)
```

```python
#importing package for classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#split train into features and target column
X, y = train.values[:, :-1], train.values[:, -1]
```

```python
# ensure all data are floating point values
X = X.astype('float32')
```

```python
#encode strings to integer
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
# split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# determine the number of input features
n_features = X_train.shape[1]
n_features
```

```python
# define model
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(n_features,)))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#define early_stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=15)

# fit the model
model.fit(X_train, 
          y_train, 
          batch_size = 200, 
          epochs = 100, 
          callbacks = [early_stopping],
          validation_data = (X_test, y_test), 
          verbose=2)
```

```python
# evaluate the model
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print('Test Accuracy: %.3f' % acc)
```

```python
prediction = model.predict(test)
prediction
```

```python
max_prediction = [np.argmax(prediction[i]) for i in range(len(prediction))]
```

```python
bacteria = le.inverse_transform(max_prediction)
bacteria
```

```python
tabular = pd.DataFrame({'row_id':sample_sub.row_id, 'target': bacteria})
tabular
```
